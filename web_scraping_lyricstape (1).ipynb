# Getting all the years

````python
years = soup.find_all("h4", {"class": "panel-title"})
````

# Getting all the movie links in all years

````python
divs_with_movie_links = soup.find_all("div", {"class": "panel-body"})
````

````python
# for div in divs_with_movie_links:
#   print(div)
#   break
````

````python
years_list = []
year_with_links = {}

for item in years:
  years_list.append(item.text)

for i in range(len(years_list)):
  # print(years_list[i])
  links = [a["href"] for a in divs_with_movie_links[i].select("a[href]")]
  # print(links)
  year_with_links[years_list[i]] = links
  text = [link.split('/')[-2] for link in links]
  # print(text)
  for j in range(len(links)):
    year_with_links[years_list[i]] = {'name': text[j], 'link': links[j]}

pprint.pprint(year_with_links)
````

````python
import os

def save_data_to_folder(data, folder_path, file_name):
    # Create the folder if it doesn't exist
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    # Create the file path
    file_path = os.path.join(folder_path, file_name)

    # Save the data to the file
    with open(file_path, 'w') as file:
        file.write(data)
````

````python
import re

for year in year_with_links.keys():

  for i in range(len(year_with_links[year])):
    # further link
    link=year_with_links[year]['link']

    name = year_with_links[year]['name']


    # call get method to request the page
    page=requests.get(link)
    page_content = BeautifulSoup(page.content, "html.parser")

    # Define the folder path where you want to save the data
    folder_path = f'lyricstape/{year}'

    # Define the file name
    file_name = f'{name}.html'

    # Save the data to the folder
    save_data_to_folder(str(page_content), folder_path, file_name)
````
